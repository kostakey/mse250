```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename1 = "data/S1_Al2024_11.csv"
filename2 = "data/S1_Al6061_11.csv"
filename3 = "data/S1_Brass_11.csv"
filename4 = "data/S1_Steel_11.csv"

df1 = pd.read_csv(filename1, skiprows=0, delimiter=',')
df2 = pd.read_csv(filename2, skiprows=0, delimiter=',')
df3 = pd.read_csv(filename3, skiprows=0, delimiter=',')
df4 = pd.read_csv(filename4, skiprows=0, delimiter=',')

dataset1_init_params = df1.iloc[2]
dataset2_init_params = df2.iloc[2]
dataset3_init_params = df3.iloc[2]
dataset4_init_params = df4.iloc[2]

width1_mm = float(dataset1_init_params.name[2])
thickness1_mm = float(dataset1_init_params.name[3])
length1_mm = float(dataset1_init_params.name[4])

width2_mm = float(dataset2_init_params.name[2])
thickness2_mm = float(dataset2_init_params.name[3])
length2_mm = float(dataset2_init_params.name[4])

width3_mm = float(dataset3_init_params.name[2])
thickness3_mm = float(dataset3_init_params.name[3])
length3_mm = float(dataset3_init_params.name[4])

width4_mm = float(dataset4_init_params.name[2])
thickness4_mm = float(dataset4_init_params.name[3])
length4_mm = float(dataset4_init_params.name[4])

# print(dataset1_init_params)
# print(float(width1_mm))
```

```python
# read the file but skip the beginning info
df1 = pd.read_csv(filename1, skiprows=6, delimiter=',')
df2 = pd.read_csv(filename2, skiprows=6, delimiter=',')
df3 = pd.read_csv(filename3, skiprows=6, delimiter=',')
df4 = pd.read_csv(filename4, skiprows=6, delimiter=',')

# grabs the table number
dataset1_nums = df1.iloc[:,0]
dataset2_nums = df2.iloc[:,0]
dataset3_nums = df3.iloc[:,0]
dataset4_nums = df4.iloc[:,0]

# redefine the amount of tables in the file if needed
for row_num, table in enumerate(dataset1_nums):
    if table == 1:
        num11 = row_num
    if table == 2:
        num12 = row_num
    if table == 3:
        num13 = row_num
    if table == 4:
        num14 = row_num

for row_num, table in enumerate(dataset2_nums):
    if table == 1:
        num21 = row_num
    if table == 2:
        num22 = row_num
    if table == 3:
        num23 = row_num
    if table == 4:
        num24 = row_num

for row_num, table in enumerate(dataset3_nums):
    if table == 1:
        num31 = row_num
    if table == 2:
        num32 = row_num
    if table == 3:
        num33 = row_num
    if table == 4:
        num34 = row_num

for row_num, table in enumerate(dataset4_nums):
    if table == 1:
        num41 = row_num
    if table == 2:
        num42 = row_num
    if table == 3:
        num43 = row_num
    if table == 4:
        num44 = row_num
```

```python
# redefine the amount of tables in the file if needed
time11_s = df1.iloc[num11+2:num12,1].astype(float)
displacement11_mm = df1.iloc[num11+2:num12,2].astype(float)
force11_N = df1.iloc[num11+2:num12,3].astype(float)

time21_s = df2.iloc[num21+2:num22,1].astype(float)
displacement21_mm = df2.iloc[num21+2:num22,2].astype(float)
force21_N = df2.iloc[num21+2:num22,3].astype(float)

time31_s = df3.iloc[num31+2:num32,1].astype(float)
displacement31_mm = df3.iloc[num31+2:num32,2].astype(float)
force31_N = df3.iloc[num31+2:num32,3].astype(float)

time41_s = df4.iloc[num41+2:num42,1].astype(float)
displacement41_mm = df4.iloc[num41+2:num42,2].astype(float)
force41_N = df4.iloc[num41+2:num42,3].astype(float)

## looking at the time-stops for al2024
time12_s = df1.iloc[num12+2:num13,1].astype(float)
displacement12_mm = df1.iloc[num12+2:num13,2].astype(float)
force12_N = df1.iloc[num12+2:num13,3].astype(float)

time13_s = df1.iloc[num13+2:num14,1].astype(float)
displacement13_mm = df1.iloc[num13+2:num14,2].astype(float)
force13_N = df1.iloc[num13+2:num14,3].astype(float)

time14_s = df1.iloc[num14+2:,1].astype(float)
displacement14_mm = df1.iloc[num14+2:,2].astype(float)
force14_N = df1.iloc[num14+2:,3].astype(float)
```

```python
area1 = width1_mm * thickness1_mm # area of the sample in mm^2
area2 = width2_mm * thickness2_mm # area of the sample in mm^2
area3 = width3_mm * thickness3_mm # area of the sample in mm^2
area4 = width4_mm * thickness4_mm # area of the sample in mm^2
```

```python
force11_kN = force11_N * .001 # N to kN
force21_kN = force21_N * .001 # N to kN
force31_kN = force31_N * .001 # N to kN
force41_kN = force41_N * .001 # N to kN

force12_kN = force12_N * .001
force13_kN = force13_N * .001
force14_kN = force14_N * .001
```

```python
strain11 = displacement11_mm / length1_mm # engineering strain
stress11 = (force11_kN / area1) * 1000 # engineering stress in MPa

strain2 = displacement21_mm / length2_mm # engineering strain
stress2 = (force21_kN / area2) * 1000 # engineering stress in MPa

strain3 = displacement31_mm / length3_mm # engineering strain
stress3 = (force31_kN / area3) * 1000 # engineering stress in MPa

strain4 = displacement41_mm / length4_mm # engineering strain
stress4 = (force41_kN / area4) * 1000 # engineering stress in MPa


strain12 = displacement12_mm / length1_mm
stress12 = (force12_kN / area1) * 1000

strain13 = displacement13_mm / length1_mm
stress13 = (force13_kN / area1) * 1000

strain14 = displacement14_mm / length1_mm
stress14 = (force14_kN / area1) * 1000
```

```python
strain11_t = np.log(1 + strain11) # true strain
stress11_t = stress11 * (1 + strain11) # true stress

strain2_t = np.log(1 + strain2) # true strain
stress2_t = stress2 * (1 + strain2) # true stress

strain3_t = np.log(1 + strain3) # true strain
stress3_t = stress3 * (1 + strain3) # true stress

strain4_t = np.log(1 + strain4) # true strain
stress4_t = stress4 * (1 + strain4) # true stress
```

```python
## Engineering Stress vs Engineering Strain

# plt.plot(x,y, linestyle='-',marker='.',color='blue',linewidth=1)
# plt.plot(0.5*x,0.5*y,color='red')
plt.figure(figsize=(8,6))
plt.plot(np.array(strain11),np.array(stress11),color='blue')
plt.plot(np.array(strain2),np.array(stress2),color='green')
plt.plot(np.array(strain3),np.array(stress3),color='red')
plt.plot(np.array(strain4),np.array(stress4),color='black')

plt.xlabel("Engineering Strain ()")
plt.ylabel("Engineering Stress () [MPa]")
plt.title("Engineering Stress vs. Engineering Strain")

# strain_max = np.max(strain4)
# plt.xticks(np.arange(0,strain_max,0.05))
plt.xticks(np.arange(0,0.5,0.05))

# stress_max = np.max(stress1)
# plt.yticks(np.arange(0,stress_max,50))
plt.yticks(np.arange(0,550,50))

plt.legend(["Al2024","Al6061","Brass","Steel"])
plt.grid()
```

```python
## True Stress vs. True Strain

plt.figure(figsize=(8,6))
plt.plot(np.array(strain11_t),np.array(stress11_t),color='blue')
plt.plot(np.array(strain2_t),np.array(stress2_t),color='green')
plt.plot(np.array(strain3_t),np.array(stress3_t),color='red')
plt.plot(np.array(strain4_t),np.array(stress4_t),color='black')

plt.xlabel("True Strain ()")
plt.ylabel("True Stress () [MPa]")
plt.title("True Stress vs. True Strain")

plt.xticks(np.arange(0,0.45,0.05))
plt.yticks(np.arange(0,650,50))

plt.legend(["Al2024","Al6061","Brass","Steel"])
plt.grid()
```
