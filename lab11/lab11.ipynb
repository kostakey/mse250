```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the carbon content for each steel type
carbon_content = {
    '1018': 0.18,
    '1045': 0.45,
    '1070': 0.70,
    '1095': 0.95
}

# Heat treatment labels in order of increasing cooling rate
heat_treatments = ['FC', 'AC', 'OQ', 'WQ', 'TP']
cooling_rate_labels = ['FC (slow)', 'AC', 'OQ', 'WQ (fast)', 'TP']

# Hardness data extracted from the image for each treatment and steel type
hardness_data = {
    '1018': {
        'FC': [-17, -17, -16],
        'AC': [-16, -9.5, -8.5],
        'OQ': [1.5, 5.5, 6.5],
        'WQ': [34, 27, 22],
        'TP': [17, 13.5, 16]
    },
    '1045': {
        'FC': [-5, -6, -3],
        'AC': [6, 3, 5],
        'OQ': [10, 12, 17],
        'WQ': [22, 29, 37],
        'TP': [11, 15, 16]
    },
    '1070': {
        'FC': [15.5, 20, 16],
        'AC': [59.0, 56.0, 55.0],
        'OQ': [58.5, 58, 53],
        'WQ': [64.0, 63.5, 65.0],
        'TP': [38.0, 36.0, 36.0]
    },
    '1095': {
        'FC': [25, 24.5, 24.5],
        'AC': [42.5, 43.5, 44],
        'OQ': [55, 54.5, 58],
        'WQ': [56.5, 56.8, 58],
        'TP': [40.5, 40, 40.5]
    }
}
```

```python
# Compute average hardness for each heat treatment and steel type
avg_hardness_by_treatment = {
    steel: [np.mean(hardness_data[steel][treatment]) for treatment in heat_treatments]
    for steel in hardness_data
}
```

```python
# First plot: Hardness vs. Carbon Content for each heat treatment
plt.figure(figsize=(10, 6))
for i, treatment in enumerate(heat_treatments):
    avg_hardness = [np.mean(hardness_data[steel][treatment]) for steel in carbon_content]
    plt.plot(list(carbon_content.values()), avg_hardness, marker='o', label=treatment)
plt.xlabel("Carbon Content (%)")
plt.ylabel("Average Hardness")
plt.title("Hardness vs. Carbon Content for Different Heat Treatments")
plt.legend(title="Heat Treatment")
# plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Second plot: Hardness vs. Cooling Rate for each steel type
plt.figure(figsize=(10, 6))
x = np.arange(len(heat_treatments))
for steel in carbon_content:
    plt.plot(x, avg_hardness_by_treatment[steel], marker='o', label=steel)
plt.xticks(x, cooling_rate_labels)
plt.xlabel("Cooling Rate (Heat Treatment)")
plt.ylabel("Average Hardness")
plt.title("Hardness vs. Cooling Rate for Different Steel Compositions")
plt.legend(title="Steel Type")
# plt.grid(True)
plt.tight_layout()
plt.show()
```
