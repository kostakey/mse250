```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "S1_Al6061_11.csv"

# read the file but skip the beginning info
df = pd.read_csv(filename, skiprows=6, delimiter=',')

# grabs the table number
dataset_nums = df.iloc[:,0]

# redefine the amount of tables in the file if needed
for row_num, table in enumerate(dataset_nums):
    if table == 1:
        num1 = row_num
    if table == 2:
        num2 = row_num
    if table == 3:
        num3 = row_num
    if table == 4:
        num4 = row_num
```

```python
# redefine the amount of tables in the file if needed
time1 = df.iloc[num1+2:num2,1].astype(float)
displacement1 = df.iloc[num1+2:num2,2].astype(float)
force1 = df.iloc[num1+2:num2,3].astype(float)

time2 = df.iloc[num2+2:num3,1].astype(float)
displacement2 = df.iloc[num2+2:num3,2].astype(float)
force2 = df.iloc[num2+2:num3,3].astype(float)

time3 = df.iloc[num3+2:num4,1].astype(float)
displacement3 = df.iloc[num3+2:num4,2].astype(float)
force3 = df.iloc[num3+2:num4,3].astype(float)

time4 = df.iloc[num4+2:,1].astype(float)
displacement4 = df.iloc[num4+2:,2].astype(float)
force4 = df.iloc[num4+2:,3].astype(float)
```

```python
# x = data.iloc[0:1,1]
# width = .508 * 25.4 # width in mm
# thickness = .062 * 25.4 # thickness in mm
# length = 2.495 * 25.4 # length in mm

force1_kN = force1 * .001 # N to kN
y1 = force1_kN
displacement1_inch = displacement1 * 0.0394 # mm to inch
x1 = displacement1_inch

force2_kN = force2 * .001 # N to kN
y2 = force2_kN
displacement2_inch = displacement2 * 0.0394 # mm to inch
x2 = displacement2_inch

force3_kN = force3 * .001 # N to kN
y3 = force3_kN
displacement3_inch = displacement3 * 0.0394 # mm to inch
x3 = displacement3_inch

force4_kN = force4 * .001 # N to kN
y4 = force4_kN
displacement4_inch = displacement4 * 0.0394 # mm to inch
x4 = displacement4_inch
```

```python
## Just Engineering Stress vs Strain

# plt.plot(x,y, linestyle='-',marker='.',color='blue',linewidth=1)
# plt.plot(0.5*x,0.5*y,color='red')
plt.plot(x1,y1,color='blue')

plt.xlabel("Displacement (in)")
plt.ylabel("Force (kN)")

x1_max = np.max(x1)
plt.xticks(np.arange(0,x1_max,0.1))

y1_max = np.max(y1)
plt.yticks(np.arange(0,y1_max,1))

plt.legend(["Al6061","Al6061_1"])
```
