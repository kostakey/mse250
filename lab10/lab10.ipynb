```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example data (replace with your actual yield/flow stresses and rates)
deformation_rates_hdpe = [1, 5, 20]  # mm/s
yield_hdpe = [124, 139, 155]  # MPa
flow_hdpe = [87, 95, 104]     # MPa

deformation_rates_ldpe = [1, 5, 20]
yield_ldpe = [52, 55, 60]
flow_ldpe = [47, 49, 51]

deformation_rates_pc = [1, 5, 20]
yield_pc = [288, 298, 310]
flow_pc = [229, 240, 248]

deformation_rates_pmma = [0.1, 1]  # e.g., 0.1 in/min and 1 in/min
yield_pmma_rt = [65, 70]  # At room temp
flow_pmma_rt = [55, 60]

yield_pmma_50c = [50, 55]  # At 50 C
flow_pmma_50c = [43, 48]
```

```python
plt.figure(figsize=(8, 6))

plt.plot(deformation_rates_hdpe, yield_hdpe, 'o-', label='HDPE')
plt.plot(deformation_rates_ldpe, yield_ldpe, 'o-', label='LDPE')
plt.plot(deformation_rates_pc, yield_pc, 'o-', label='PC')
plt.plot(deformation_rates_pmma, yield_pmma_rt, 'o-', label='PMMA (RT)')
plt.plot(deformation_rates_pmma, yield_pmma_50c, 'o-', label='PMMA (50°C)')

plt.xscale('log')
plt.xlabel('Deformation Rate [in/min]')
plt.ylabel('Yield Stress [MPa]')
plt.title('Yield Stress vs. Deformation Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.plot(deformation_rates_hdpe, flow_hdpe, 'o-', label='HDPE')
plt.plot(deformation_rates_ldpe, flow_ldpe, 'o-', label='LDPE')
plt.plot(deformation_rates_pc, flow_pc, 'o-', label='PC')
plt.plot(deformation_rates_pmma, flow_pmma_rt, 'o-', label='PMMA (RT)')
plt.plot(deformation_rates_pmma, flow_pmma_50c, 'o-', label='PMMA (50°C)')

plt.xscale('log')
plt.xlabel('Deformation Rate [in/min]')
plt.ylabel('Flow Stress [MPa]')
plt.title('Flow Stress vs. Deformation Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
